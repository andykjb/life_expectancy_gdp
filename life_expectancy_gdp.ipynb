```python
df = df.rename(columns={'Life expectancy at birth (years)': 'life_ex'})
```

#Exploration

```python
g = sns.FacetGrid(df, col="Country", 
                  col_wrap=2, 
                  sharex=False, 
                  sharey=False,
                  height=4,
                  aspect=1.5)
g.map(sns.lineplot, "Year","life_ex")
plt.show()
```

Plotting the data over years we can see that life expectancy goes up from year 2000 to year 2014, for all countries. 

Zimbabwe dips below initial life expectancy during the period 2001 - 2006 but goes up after that.



```python
g = sns.FacetGrid(df, col="Country", 
                  col_wrap=2, 
                  sharex=False, 
                  sharey=False,
                  height=4,
                  aspect=1.5)
g.map(sns.lineplot, "Year","GDP")
plt.show()
```

Plotting GDP we can see that all countries seem to have an increase in GDP over the years. 
Only China and US seem to have the most steady increase in GDP over time.

```python
sns.set_context("notebook")

sns.lmplot(data=df, 
           y='GDP', 
           x='life_ex', 
           col='Country', 
           col_wrap=2, 
           facet_kws=dict(sharey=False, sharex=False))
```

By comparing GDP and Life Expectancy we can see that they are clearly tied together in each country.
When GDP goes up, Life Expectancy goes up as well as the other way around

```python
fig, ax = plt.subplots(figsize=(14,10))
ax= sns.boxplot(data=df[df.Country != 'Zimbabwe'],
            y='life_ex',
            x='Country')


```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax= sns.boxplot(data=df[df.Country == 'Zimbabwe'],
            y='life_ex',
            x='Country')
```

With Zimbabwe as an outlier it is easier to get a more accurate visualisation of life expectancy in the different countries by having Zimbabwe in its own boxplot.

```python
fig, ax = plt.subplots(figsize=(14,10))
ax = sns.violinplot(data=df[df.Country != 'Zimbabwe'],
            y='life_ex',
            x='Country')
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax = sns.violinplot(data=df[df.Country == 'Zimbabwe'],
            y='life_ex',
            x='Country')
```

Doing the same breakout of Zimbabwe with violinplots, we are better able to note the distribution of life expectancy in the different countries
