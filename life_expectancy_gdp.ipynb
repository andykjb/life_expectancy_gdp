```python
df = df.rename(columns={'Life expectancy at birth (years)': 'life_ex'})
```

#Exploration

```python
fig, ax = plt.subplots(figsize=(12,10))


ax = sns.lineplot(data=df, x='Year', y='life_ex', hue='Country',
                  legend='full',
                  palette='tab10')

sns.set_style("whitegrid")
ax.set(title='Life Expectancy YoY', ylabel='Life expectancy')
```

Plotting the data over years we can see that life expectancy goes up from year 2000 to year 2014, for all countries. 

Zimbabwe dips below initial life expectancy during the period 2001 - 2006 but goes up after that.



```python
fig, ax = plt.subplots(figsize=(12,10))


ax = sns.lineplot(data=df, x='Year', y='GDP', hue='Country',
                  legend='full',
                  palette='tab10')

sns.set_style("whitegrid")
ax.set(title='GDP YoY')
```
